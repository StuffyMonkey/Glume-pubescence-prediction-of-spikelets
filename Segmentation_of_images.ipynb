```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
from matplotlib import image as mpimg
import os
import re

h_dir = os.getcwd()
```

```python
os.chdir('/home/rostepifanov/import/bin/segmentation')
save_dir = '/home/jupyter-n.artemenko/projects/spikelet_pubescence/masks/'

seg_script = """./infer -bone efficientnet-b2 -mn model_efficientnet-b2.bin --cuda --verbose -bs 32 -ip {img_path} -op {out_path} """

ploid_patt = r'([А-Яа-я]+плоид[а-я]+)'
pin_patt = r'[\d]{2,5}_[^,(I|l)]_\{[XХI]+-?\d{2}\}\.(jpg|JPG)$'
table_patt = r'[\d]{2,5}_\{[ХXI]+-?\d{2}\}\.(jpg|JPG)$'
```

## Get masks of images for the main task

```python
for dirpath, dirnames, filenames in os.walk('/data/cv_project/spikedroid/опушение'):
    for fn in filenames:           
            if re.search(pin_patt, fn):
                tmp = re.search(pin_patt, fn).group()
                ploid_res = re.search(ploid_patt, dirpath).group(1)
                if not tmp.replace('.jpg', '.png') in os.listdir('/home/jupyter-n.artemenko/projects/spikelet_pubescence/masks/' + ploid_res + '/pin') and not tmp.replace('.JPG', '.png') in os.listdir('/home/jupyter-n.artemenko/projects/spikelet_pubescence/masks/' + ploid_res + '/pin'):
                    os.system(seg_script.format(img_path= (dirpath+'/'+tmp).replace(' ', '\ '), 
                            out_path=h_dir + '/masks/'+ploid_res+'/pin'))
                
            if re.search(table_patt, fn):
                tmp = re.search(table_patt, fn).group()
                ploid_res = re.search(ploid_patt, dirpath).group(1)
                if not tmp.replace('.jpg', '.png') in os.listdir('/home/jupyter-n.artemenko/projects/spikelet_pubescence/masks/' + ploid_res + '/table') and not tmp.replace('.JPG', '.png') in os.listdir('/home/jupyter-n.artemenko/projects/spikelet_pubescence/masks/' + ploid_res + '/table'):
                    os.system(seg_script.format(img_path= (dirpath+'/'+tmp).replace(' ', '\ '), 
                            out_path=h_dir + '/masks/'+ploid_res+'/table'))
```

## Masks for holdout

```python
save_dir = '/home/jupyter-n.artemenko/projects/spikelet_pubescence/test/Holdout/masks/'
holdout_dir = '/data/cv_project/spikedroid/опушение/'

seg_script = './infer -bone efficientnet-b2 -mn model_efficientnet-b2.bin --cuda --verbose -bs 32 -ip {img_path} -op {out_path}'
segmentation = lambda x,y: os.system(seg_script.format(img_path=x, out_path=y))

for dirpath, dirname, filenames in os.walk(holdout_dir):
    if 'опушением' in dirpath:
        for fn in filenames:
            img_path = f'{dirpath}/{fn}'.replace(' ', '\ ')
            out_path = f'{save_dir}неопушенные'.replace(' ', '\ ')
            if not os.path.exists(f'{save_dir}опушенные'):
                os.mkdir(f'{save_dir}опушенные')
            segmentation(img_path, out_path)
    else:
        for fn in filenames:
            out_path = f'{save_dir}опушенные'.replace(' ', '\ ')
            img_path =  f'{dirpath}/{fn}'.replace(' ', '\ ')
            if not os.path.exists(f'{save_dir}неопушенные'):
                os.mkdir(f'{save_dir}неопушенные')
            segmentation(img_path, out_path)
```
